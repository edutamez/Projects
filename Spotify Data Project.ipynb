```python
#finding top 10 tracks by querying by popularity and sorting in ascending fasion
top10 = data_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
top10[:10]
```

```python
#find the 10 least popular songs in the data set
worst10 = data_tracks.sort_values('popularity', ascending = True).head(10)
worst10
```

```python
data_tracks.set_index('release_date',inplace = True)
data_tracks.index = pd.to_datetime(data_tracks.index)
data_tracks.head()
```

```python
#checks artist at 18th row
data_tracks[['artists']].iloc[18]
```

```python
#changes duration value from seconds to milliseconds
data_tracks['duration'] = data_tracks['duration_ms'].apply(lambda x : round(x/1000))
data_tracks.drop('duration_ms', inplace = True, axis = 1)
data_tracks.duration.head()
```

```python
cm = data_tracks.drop(['key','mode','explicit'], axis = 1).corr(method = 'pearson')
plt.figure(figsize=(14,6),facecolor='blue')
map = sns.heatmap(cm, annot= True,fmt='.1g',vmin=-1,vmax=1,center=0,cmap='inferno',linewidths=1,linecolor='Black')
map.set_title('Correlattion Heatmap between Variable')
map.set_xticklabels(map.get_xticklabels(), rotation=90)
```

```python
# sets parameter for how much of the dataset to use, in this case we use 0.4%
sam = data_tracks.sample(int(0.004 * len(data_tracks)))
len(sam)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='loudness', x = 'energy',scatter_kws={"color": "cyan"}, line_kws={"color": "blue"}).set(title='Loudness vs Energy Correlation')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sam, y='popularity', x='acousticness', color='b').set(title='Popularity vs Acousticness Correlation')
```
